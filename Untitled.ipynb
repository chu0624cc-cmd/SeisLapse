```python
import numpy as np
import obspy
from obspy import Trace, Stream, UTCDateTime
from scipy.signal import fftconvolve, butter, filtfilt, hilbert
from scipy.interpolate import interp1d
from scipy.fft import rfft, irfft, rfftfreq, next_fast_len
import os

FS_RAW     = 100.0
DURATION   = 86400.0
TRUE_DVV   = 0.1e-2
START_JDAY = 280
OUTDIR     = './synthetic/2019'
N_DAYS     = 30

dvv_true = np.zeros(N_DAYS)
dvv_true[10:15] = +TRUE_DVV
dvv_true[15:20] =  0.0
dvv_true[20:30] = -TRUE_DVV

np.random.seed(2025)
os.makedirs(OUTDIR, exist_ok=True)

# ── 格林函数（多频段，模拟真实散射）────────────────────────────
def make_green(fs, t_max=200.0):
    N   = int(t_max * fs)
    t   = np.arange(N) / fs
    rng = np.random.default_rng(42)
    g   = np.zeros(N)
    t0  = 3.0

    # 各频段成分
    components = [
        # (freq_center, t_decay, amplitude, label)
        (0.07,  80.0, 0.3,  'single_freq_microseism'),
        (0.15,  60.0, 1.0,  'double_freq_microseism'),   # 主要成分
        (1.5,   25.0, 0.6,  'target_1_2Hz'),             # dv/v目标频段
        (7.0,    5.0, 0.2,  'high_freq_scatter'),
    ]

    for freq_c, t_decay, amp, label in components:
        # 直达波
        g += amp * np.exp(-((t-t0)**2)/(0.3**2)) * \
             np.sin(2*np.pi*freq_c*(t-t0))

        # 散射尾波
        for tau in np.arange(t0*2, min(t_max*0.8, 80.0), 0.5):
            a     = amp * np.exp(-(tau-t0)/t_decay)
            phase = rng.uniform(0, 2*np.pi)
            bw    = rng.uniform(0.2, 0.6)
            g    += a * 0.15 * np.cos(2*np.pi*freq_c*t+phase) * \
                    np.exp(-((t-tau)**2)/(bw**2))

    g /= np.max(np.abs(g)) + 1e-30
    return t, g

t_g, green_ref = make_green(FS_RAW)
t_orig = np.arange(len(green_ref)) / FS_RAW

def make_green_dvv(dvv):
    if dvv == 0.0:
        return green_ref.copy()
    scale = 1.0 + dvv
    return interp1d(t_orig, green_ref,
                    bounds_error=False, fill_value=0.0)(t_orig * scale)

# ── 真实频段结构的噪声源 ──────────────────────────────────────────
def make_colored_noise(N, fs, day_idx=0):
    """
    模拟真实地震环境噪声频谱：
    - 双频微震主峰（0.1~0.2Hz）
    - 人文噪声日变化（1~2Hz，白天强）
    - 高频衰减（>10Hz）
    """
    Nfft  = next_fast_len(N)
    freqs = rfftfreq(Nfft, d=1.0/fs)
    freqs[0] = 1e-10   # 避免除零

    # 基础功率谱（频率的负幂次，模拟1/f^2背景）
    psd = 1.0 / (freqs**2 + 1e-10)

    # 双频微震峰值（0.1~0.2Hz）
    f_dfm = 0.15
    psd  += 8.0 * np.exp(-((freqs - f_dfm)**2) / (0.03**2))

    # 单频微震（0.05~0.1Hz）
    f_sfm = 0.07
    psd  += 3.0 * np.exp(-((freqs - f_sfm)**2) / (0.01**2))

    # 人文噪声（1~2Hz）：白天强（振幅×3），夜晚弱（振幅×1）
    # day_idx控制日变化：奇数天白天噪声强
    human_amp = 3.0 if (day_idx % 2 == 0) else 1.0
    psd      += human_amp * np.exp(-((freqs - 1.5)**2) / (0.3**2))

    # 高频截断（>20Hz）
    psd *= np.exp(-(freqs / 20.0)**4)

    # 生成带颜色的随机噪声
    white  = np.random.randn(Nfft) + 1j * np.random.randn(Nfft)
    X      = white[:len(freqs)] * np.sqrt(psd)
    noise  = np.real(irfft(X, Nfft))[:N]
    noise /= np.std(noise) + 1e-30
    return noise

# ── 偶发事件（模拟地震、爆破）────────────────────────────────────
def add_transients(sig, fs, n_events=3):
    """
    在随机时刻加入短时高振幅瞬态（模拟地震/爆破）
    → 测试你的MAX_STD质量控制是否能剔除
    """
    N   = len(sig)
    out = sig.copy()
    for _ in range(n_events):
        t_start = np.random.randint(0, N - int(30*fs))
        duration = int(np.random.uniform(5, 30) * fs)
        amp      = np.random.uniform(5, 15)   # 振幅是正常的5~15倍
        burst    = amp * np.random.randn(duration) * \
                   np.hanning(duration)
        out[t_start:t_start+duration] += burst
    return out

# ── 生成1天连续波形 ───────────────────────────────────────────────
def make_day_signal(fs, duration, green, day_idx=0,
                    add_events=True, n_events=3):
    N      = int(duration * fs)

    # 有色噪声源（多频段）
    source = make_colored_noise(N, fs, day_idx=day_idx)

    # 卷积格林函数
    sig    = fftconvolve(source, green, mode='full')[:N]
    sig   /= np.std(sig) + 1e-30

    # 加偶发瞬态事件
    if add_events:
        sig = add_transients(sig, fs, n_events=n_events)

    return sig.astype(np.float64)

# ── 生成30天SAC ───────────────────────────────────────────────────
print(f"生成 {N_DAYS} 天合成SAC（真实频段结构）\n")
print(f"{'Day':<6} {'jday':<6} {'真实dv/v':>10}  干扰设计")
print("─" * 60)

for d in range(N_DAYS):
    jday  = START_JDAY + d
    dvv_d = dvv_true[d]
    green = make_green_dvv(dvv_d)

    # 每天有不同的干扰强度
    n_ev  = np.random.randint(2, 6)   # 2~5个偶发事件
    sig   = make_day_signal(FS_RAW, DURATION, green,
                            day_idx=d, add_events=True, n_events=n_ev)

    tr                     = Trace()
    tr.data                = sig
    tr.stats.sampling_rate = FS_RAW
    tr.stats.network       = 'AL'
    tr.stats.station       = 'AL201'
    tr.stats.channel       = 'BHZ'
    tr.stats.starttime     = UTCDateTime('2019-01-01') + (jday-1)*86400

    fname = os.path.join(OUTDIR, f'AL201_Z_{jday}.sac')
    Stream([tr]).write(fname, format='SAC')

    # 段落标签
    if d < 10:   seg = 'baseline   (dv/v= 0.00%)'
    elif d < 15: seg = 'increase   (dv/v=+0.10%)'
    elif d < 20: seg = 'recovery   (dv/v= 0.00%)'
    else:        seg = 'decrease   (dv/v=-0.10%)'

    human = '人文噪声强' if d%2==0 else '人文噪声弱'
    print(f"Day{d+1:<3} jday={jday}  {dvv_d*100:>+8.4f}%  "
          f"{seg}  事件={n_ev}个  {human}")

# ── 频谱可视化验证 ────────────────────────────────────────────────
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('合成数据频段特性验证', fontweight='bold')

# (a) 格林函数频谱
ax = axes[0, 0]
Nfft   = next_fast_len(len(green_ref))
freqs  = rfftfreq(Nfft, d=1.0/FS_RAW)
G      = np.abs(rfft(green_ref, Nfft))
ax.semilogy(freqs[1:], G[1:], lw=1.0, color='k')
for f, label in [(0.07,'SFM'),(0.15,'DFM'),(1.5,'Target'),(7.0,'HF')]:
    ax.axvline(f, ls='--', lw=0.8, alpha=0.7, label=label)
ax.set_xlim(0.01, 25)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Amplitude (log)')
ax.set_title('(a) 格林函数频谱')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.3)

# (b) 噪声源频谱（白天vs夜晚）
ax = axes[0, 1]
N_demo = int(3600 * FS_RAW)   # 1小时
for di, label, col in [(0,'白天(偶数天)','tomato'),
                        (1,'夜晚(奇数天)','steelblue')]:
    noise = make_colored_noise(N_demo, FS_RAW, day_idx=di)
    Nf    = next_fast_len(N_demo)
    fq    = rfftfreq(Nf, d=1.0/FS_RAW)
    sp    = np.abs(rfft(noise, Nf))
    ax.semilogy(fq[1:], sp[1:], lw=1.0, color=col, alpha=0.8, label=label)
ax.axvspan(1.0, 2.0, alpha=0.1, color='green', label='目标频带1~2Hz')
ax.set_xlim(0.01, 25)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Amplitude (log)')
ax.set_title('(b) 噪声源频谱（含日变化）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.3)

# (c) 合成波形频谱（基准天 vs 变化天）
ax = axes[1, 0]
np.random.seed(42)
for dvv_d, label, col in [
    (0.0,      'baseline v0',         'k'),
    (+TRUE_DVV, f'v0+{TRUE_DVV*100:.2f}%', 'tomato'),
    (-TRUE_DVV, f'v0-{TRUE_DVV*100:.2f}%', 'steelblue'),
]:
    g_d  = make_green_dvv(dvv_d)
    sig  = make_day_signal(FS_RAW, 3600.0, g_d, day_idx=0, add_events=False)
    Nf   = next_fast_len(len(sig))
    fq   = rfftfreq(Nf, d=1.0/FS_RAW)
    sp   = np.abs(rfft(sig, Nf))
    ax.semilogy(fq[1:], sp[1:], lw=0.9, color=col, alpha=0.8, label=label)
ax.axvspan(1.0, 2.0, alpha=0.1, color='green', label='目标频带')
ax.set_xlim(0.01, 25)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Amplitude (log)')
ax.set_title('(c) 合成波形频谱（三种速度）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.3)

# (d) 含瞬态事件的波形片段
ax = axes[1, 1]
np.random.seed(2025)
sig_clean = make_day_signal(FS_RAW, 3600.0, green_ref,
                             day_idx=0, add_events=False)
sig_noisy = make_day_signal(FS_RAW, 3600.0, green_ref,
                             day_idx=0, add_events=True, n_events=3)
t_seg = np.arange(int(600*FS_RAW)) / FS_RAW   # 前600s
ax.plot(t_seg, sig_clean[:len(t_seg)]/np.std(sig_clean) + 5,
        lw=0.5, color='steelblue', label='无瞬态事件')
ax.plot(t_seg, sig_noisy[:len(t_seg)]/np.std(sig_noisy),
        lw=0.5, color='tomato',    label='含瞬态事件（地震/爆破）')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Norm. Amplitude')
ax.set_title('(d) 波形对比（含/不含瞬态干扰）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.4)

plt.tight_layout()
plt.savefig('synthetic_spectrum_check.png', dpi=130, bbox_inches='tight')
plt.show()

print(f"\n合成数据设计汇总：")
print(f"  格林函数频段 : 0.07Hz(SFM) + 0.15Hz(DFM) + 1.5Hz(目标) + 7Hz(高频)")
print(f"  噪声源       : 有色噪声（1/f²背景 + 各频段峰值）")
print(f"  日变化       : 偶数天人文噪声强（×3），奇数天弱（×1）")
print(f"  偶发事件     : 每天2~5个瞬态高振幅事件（振幅5~15倍）")
print(f"  速度变化     : Day1~10=0%  Day11~15=+0.1%  Day16~20=0%  Day21~30=-0.1%")
print(f"\n  这些干扰会测试你的pipeline：")
print(f"  ✓ MAX_STD质控  → 能否剔除瞬态事件窗口")
print(f"  ✓ RAM归一化    → 能否压制双频微震强能量")
print(f"  ✓ 白化         → 能否均衡各频段能量")
print(f"  ✓ dv/v         → 在有色噪声背景下能否恢复真实速度变化")
```

```python
import numpy as np
import obspy
import os
import glob
from obspy import UTCDateTime
from scipy.signal import butter, filtfilt, hilbert
import matplotlib.pyplot as plt

# ══════════════════════════════════════════════════════════════════
# Part 2：加载合成数据 → 完整pipeline → ACF
# ══════════════════════════════════════════════════════════════════
from slide      import make_raw_data
from preprocess import preprocess
from correlate  import correlate
from stacking   import stack
from process_raw import process_raw


# 恢复必要的变量
N_DAYS     = 30
N_DAYS_REF = 10
N_DAYS_CUR = 20
TRUE_DVV   = 0.1e-2

dvv_true = np.zeros(N_DAYS)
dvv_true[10:15] = +TRUE_DVV
dvv_true[15:20] =  0.0
dvv_true[20:30] = -TRUE_DVV

seg_colors = (
    ['steelblue'] * 10 +
    ['tomato']    *  5 +
    ['seagreen']  *  5 +
    ['purple']    * 10
)

def day_tag(i):
    if i < 10: return '基准'
    if i < 15: return '增大'
    if i < 20: return '恢复'
    return '减小'

print("✓ 变量恢复完成，SAC文件直接从./synthetic/2019/读取")
print(f"  N_DAYS={N_DAYS}  N_DAYS_REF={N_DAYS_REF}  TRUE_DVV={TRUE_DVV*100:.4f}%")



# ── 参数（对齐你的Cell 2，但关键区别：paz=None跳过去响应）────────
SYN_DIRS   = ['./synthetic/2019']
TARGET_SR  = 50.0
CC_LEN     = 1800
CC_STEP    = 450


FREQMIN    = 0.1
FREQMAX    = 10

MAX_STD    = 10.0
TIME_NORM  = 'ram'
RAM_HALF_WIN    = int(TARGET_SR / FREQMIN / 2)
WHITEN_METHOD   = 'smoothed'
SMOOTH_HALF_WIN = 5
PAD             = 50


WHITEN_FREQMIN  = 0.1
WHITEN_FREQMAX  = 0.1


MAXLAG          = 80.0
DO_CLEAN_UP     = True
MAX_LENGTH      = 20.0
DO_WHITEN_ACF   = False
ACF_FREQMIN     = 1.0
ACF_FREQMAX     = 2.0
ACF_WHITEN_METHOD = 'smoothed'
STACK_METHOD    = 'linear'

# ── 文件索引 ─────────────────────────────────────────────────────
file_dict = {}
for d in SYN_DIRS:
    for fp in glob.glob(os.path.join(d, '*_Z_*.sac')):
        try:
            parts = os.path.basename(fp).split('_')
            jday  = int(parts[-1].split('.')[0])
            year  = int(os.path.basename(os.path.dirname(fp)))
            file_dict[(year, jday)] = fp
        except (ValueError, IndexError):
            continue
sorted_keys = sorted(file_dict)
print(f"\n找到 {len(file_dict)} 个SAC文件: {sorted_keys}")

# ── Stage A：读取 + 下采样（跳过去响应）─────────────────────────
rawdata_dict = {}
print(f"\nStage A：读取 + 下采样（不去响应）")
for year, jday in sorted_keys:
    fp = file_dict[(year, jday)]
    try:
        stream = obspy.read(fp)
        stream.merge(method=1, fill_value=0)
        trace   = stream[0]
        fs_orig = float(trace.stats.sampling_rate)
        x_raw   = trace.data.astype(np.float64)
        x_raw   = np.nan_to_num(x_raw, nan=0.0)

        st = trace.stats.starttime
        t0 = st - UTCDateTime(st.year, st.month, st.day,
                               st.hour, st.minute, st.second)

        # ← paz=None 跳过去响应，合成数据不需要
        x_proc = process_raw(
            x=x_raw, fs_old=fs_orig, fs_new=TARGET_SR,
            t0=t0, phi_shift=False,
            paz=None, pre_filt=None,
        )

        raw = make_raw_data(
            x=x_proc, fs=TARGET_SR, cc_len=CC_LEN, cc_step=CC_STEP,
            starttime=float(trace.stats.starttime),
            endtime=float(trace.stats.endtime),
            freqmin=FREQMIN, freqmax=FREQMAX,
            net=trace.stats.network, sta=trace.stats.station,
            loc=trace.stats.location, cha=trace.stats.channel,
        )

        rawdata_dict[(year, jday)] = raw
        print(f"  {year}.{jday}  windows={raw['x'].shape[1]}  "
              f"max={np.max(np.abs(raw['x'])):.3e}  ✅")
    except Exception as e:
        print(f"  {year}.{jday}  ❌ {e}")

# ── Stage B：预处理 ───────────────────────────────────────────────
preproc_dict = {}
print(f"\nStage B：预处理")
for year, jday in sorted(rawdata_dict):
    raw = rawdata_dict[(year, jday)]
    try:
        d = preprocess(
            raw=raw, freqmin=FREQMIN, freqmax=FREQMAX,
            max_std=MAX_STD,
            time_norm_method=TIME_NORM, ram_half_win=RAM_HALF_WIN,
            clip_factor=3.0,
            whiten_method=WHITEN_METHOD,
            smooth_half_win=SMOOTH_HALF_WIN, pad=PAD,
            whiten_freqmin=WHITEN_FREQMIN,
            whiten_freqmax=WHITEN_FREQMAX,
        )
        preproc_dict[(year, jday)] = d
        print(f"  {year}.{jday}  windows={d['x_white'].shape[1]}  ✅")
    except Exception as e:
        print(f"  {year}.{jday}  ❌ {e}")

# ── Stage C：相关计算 ─────────────────────────────────────────────
all_corr  = []
all_t_day = []
keys_b    = sorted(preproc_dict)
print(f"\nStage C：相关计算")
for year, jday in keys_b:
    d = preproc_dict[(year, jday)]
    try:
        corr_data = correlate(
            data1=d, maxlag=MAXLAG, data2=None,
            use_freq_domain=True,
            do_clean_up=DO_CLEAN_UP, do_abs_max=False,
            max_length=MAX_LENGTH,
            do_whiten_acf=DO_WHITEN_ACF,
            acf_freqmin=ACF_FREQMIN, acf_freqmax=ACF_FREQMAX,
            acf_whiten_method=ACF_WHITEN_METHOD,
        )
        all_corr.append(corr_data['corr'])
        all_t_day.append(corr_data['t'][0])
        lags = corr_data['lags']
        print(f"  {year}.{jday}  corr={corr_data['corr'].shape}  "
              f"max={np.max(np.abs(corr_data['corr'])):.3e}  ✅")
    except Exception as e:
        print(f"  {year}.{jday}  ❌ {e}")

# ── Stage C：相关计算 ─────────────────────────────────────────────
all_corr  = []
all_t_day = []
keys_b    = sorted(preproc_dict)
print(f"\nStage C：相关计算")
for year, jday in keys_b:
    d = preproc_dict[(year, jday)]
    try:
        corr_data = correlate(
            data1=d, maxlag=MAXLAG, data2=None,
            use_freq_domain=True,
            do_clean_up=DO_CLEAN_UP, do_abs_max=False,
            max_length=MAX_LENGTH,
            do_whiten_acf=DO_WHITEN_ACF,
            acf_freqmin=ACF_FREQMIN, acf_freqmax=ACF_FREQMAX,
            acf_whiten_method=ACF_WHITEN_METHOD,
        )
        all_corr.append(corr_data['corr'])
        all_t_day.append(corr_data['t'][0])
        lags = corr_data['lags']
        print(f"  {year}.{jday}  corr={corr_data['corr'].shape}  "
              f"max={np.max(np.abs(corr_data['corr'])):.3e}  ✅")
    except Exception as e:
        print(f"  {year}.{jday}  ❌ {e}")

# ── 叠加 ──────────────────────────────────────────────────────────
print(f"\n叠加...")
def stack_days(corr_list, method='linear'):
    all_wins  = np.concatenate(corr_list, axis=1)
    fake_dict = {
        'corr'  : all_wins,
        't'     : np.arange(all_wins.shape[1], dtype=float),
        'lags'  : lags,
        'maxlag': float(np.max(np.abs(lags))),
    }
    result = stack(fake_dict, method=method, allstack=True)
    return result['corr'][:, 0]

# 总叠加 / 参考段 / 变化段
stack_ref = stack_days(all_corr[:N_DAYS_REF], STACK_METHOD)
stack_cur = stack_days(all_corr[N_DAYS_REF:], STACK_METHOD)
stack_all = stack_days(all_corr,              STACK_METHOD)

# 逐日叠加 → daily_stacks (n_lags, n_days)
daily_stacks = np.array(
    [stack_days([c], STACK_METHOD) for c in all_corr]
).T

center_idx = len(lags) // 2
lags_pos   = lags[center_idx:]

print(f"stack_ref    shape: {stack_ref.shape}")
print(f"stack_cur    shape: {stack_cur.shape}")
print(f"stack_all    shape: {stack_all.shape}")
print(f"daily_stacks shape: {daily_stacks.shape}  "
      f"← (n_lags={daily_stacks.shape[0]}, n_days={daily_stacks.shape[1]})")
print(f"lags_pos     shape: {lags_pos.shape}  "
      f"[{lags_pos[0]:.1f} ~ {lags_pos[-1]:.1f} s]")
```

```python
from scipy.ndimage import uniform_filter1d
from scipy.signal import butter, filtfilt, hilbert

# ── 平滑函数 ──────────────────────────────────────────────────────
def smooth_acf(acf, fs, fmin, fmax, method='uniform',
               cycles=1.0, verbose=False):
    f_center   = (fmin + fmax) / 2.0
    smooth_pts = max(3, int(cycles / f_center * fs))
    if smooth_pts % 2 == 0:
        smooth_pts += 1
    if verbose:
        print(f"smooth_acf: f_center={f_center:.2f}Hz  "
              f"win={smooth_pts/fs:.3f}s  pts={smooth_pts}")

    def _smooth_1d(x):
        if method == 'uniform':
            return uniform_filter1d(x, size=smooth_pts)
        elif method == 'gaussian':
            from scipy.ndimage import gaussian_filter1d
            return gaussian_filter1d(x, sigma=smooth_pts/6.0)
        elif method == 'hanning':
            win = np.hanning(smooth_pts)
            win /= win.sum()
            return np.convolve(x, win, mode='same')
        else:
            raise ValueError(f"未知method: {method}")

    if acf.ndim == 1:
        return _smooth_1d(acf)
    out = np.zeros_like(acf)
    for i in range(acf.shape[1]):
        out[:, i] = _smooth_1d(acf[:, i])
    return out

# ── dv/v参数 ──────────────────────────────────────────────────────
FILTER_MIN = 1.0
FILTER_MAX = 2.0
DO_SMOOTH  = False    # False → 关闭平滑，对比用

center_idx = np.argmin(np.abs(lags))
lags_pos   = lags[center_idx:]

b2, a2 = butter(4, [FILTER_MIN, FILTER_MAX], btype='band', fs=TARGET_SR)

def get_pos(acf_1d, smooth=DO_SMOOTH):
    bp_  = filtfilt(b2, a2, acf_1d)
    pos  = bp_[center_idx:]
    neg  = bp_[center_idx::-1]
    sym  = (pos + neg) / 2.0
    if smooth:
        sym = smooth_acf(sym, fs=TARGET_SR,
                         fmin=FILTER_MIN, fmax=FILTER_MAX,
                         method='uniform', cycles=1.0)
    sym /= np.max(np.abs(sym)) + 1e-30
    return sym

# 打印一次平滑参数
smooth_acf(np.zeros(10), fs=TARGET_SR,
           fmin=FILTER_MIN, fmax=FILTER_MAX, verbose=True)

sym_ref = get_pos(stack_ref)   # Day1~10
sym_cur = get_pos(stack_cur)   # Day11~30
sym_all = get_pos(stack_all)   # 全部

# 逐日ACF矩阵
daily_mat = np.zeros((len(lags_pos), N_DAYS))
for i in range(N_DAYS):
    daily_mat[:, i] = get_pos(daily_stacks[:, i])

print(f"sym_ref    shape: {sym_ref.shape}")
print(f"daily_mat  shape: {daily_mat.shape}")
print(f"平滑模式: {'开启' if DO_SMOOTH else '关闭'}")

# ── 可视化 ────────────────────────────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    f'合成数据ACF验证  真实dv/v阶梯±{TRUE_DVV*100:.4f}%  '
    f'平滑={"ON" if DO_SMOOTH else "OFF"}\n'
    f'Day1~10基准 | Day11~15增大 | Day16~20恢复 | Day21~30减小',
    fontweight='bold'
)

ax = axes[0, 0]
ax.plot(lags_pos, sym_ref, lw=1.2, color='steelblue',
        label='基准叠加 Day1~10 (v0)')
ax.plot(lags_pos, sym_cur, lw=1.2, color='tomato', alpha=0.85,
        label='变化叠加 Day11~30')
ax.plot(lags_pos, sym_all, lw=1.0, color='gray', alpha=0.6, ls='--',
        label='全叠加')
ax.axvspan(30, 60, alpha=0.12, color='green', label='尾波窗30~60s')
ax.set_xlim(0, 80)
ax.set_xlabel('Lag (s)')
ax.set_title('(a) 叠加ACF对比（1~2Hz）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.4)

ax = axes[0, 1]
env_r = np.abs(hilbert(sym_ref))
env_c = np.abs(hilbert(sym_cur))
ax.semilogy(lags_pos, env_r + 1e-10, lw=1.2, color='steelblue', label='基准')
ax.semilogy(lags_pos, env_c + 1e-10, lw=1.2, color='tomato',    label='变化')
ax.axvspan(30, 60, alpha=0.12, color='green', label='尾波窗')
ax.set_xlim(0, 80)
ax.set_xlabel('Lag (s)')
ax.set_ylabel('Envelope (log)')
ax.set_title('(b) 能量衰减（1~2Hz，对数坐标）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.3)

ax = axes[1, 0]
vmax = np.percentile(np.abs(daily_mat), 98)
im   = ax.imshow(
    daily_mat.T, aspect='auto', origin='lower',
    cmap='RdBu', vmin=-vmax, vmax=vmax,
    extent=[lags_pos[0], lags_pos[-1], 0, N_DAYS]
)
for y, col in [(10,'k'),(15,'tomato'),(20,'seagreen')]:
    ax.axhline(y, color=col, ls='--', lw=1.2)
ax.axvline(30, color='g', ls='--', lw=1.0)
ax.axvline(60, color='g', ls='--', lw=1.0)
ax.set_xlabel('Lag (s)')
ax.set_ylabel('Day')
ax.set_title('(c) 逐日ACF热图（1~2Hz）')
plt.colorbar(im, ax=ax)

ax = axes[1, 1]
m = (lags_pos >= 25) & (lags_pos <= 70)
ax.plot(lags_pos[m], sym_ref[m], lw=1.5, color='steelblue',
        label='基准 Day1~10')
for sl, col, label in [
    (slice(10,15), 'tomato',
     f'增大 Day11~15 (+{TRUE_DVV*100:.2f}%)'),
    (slice(15,20), 'seagreen',
     '恢复 Day16~20 (0%)'),
    (slice(20,30), 'purple',
     f'减小 Day21~30 (-{TRUE_DVV*100:.2f}%)'),
]:
    mean_seg = np.mean(daily_mat[:, sl], axis=1)
    ax.plot(lags_pos[m], mean_seg[m], lw=1.2, color=col,
            alpha=0.85, label=label)
ax.axvspan(30, 60, alpha=0.12, color='green', label='尾波窗')
ax.set_xlabel('Lag (s)')
ax.set_title('(d) 各段平均ACF放大（25~70s）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.4)

plt.tight_layout()
plt.savefig('synthetic_acf_check.png', dpi=130, bbox_inches='tight')
plt.show()

# ── 关键数值检查 ──────────────────────────────────────────────────
print(f"\n{'='*55}")
print(f"  30~60s 基准ACF RMS : "
      f"{np.std(sym_ref[(lags_pos>=30)&(lags_pos<=60)]):.6f}")
print(f"  30~60s 变化ACF RMS : "
      f"{np.std(sym_cur[(lags_pos>=30)&(lags_pos<=60)]):.6f}")
print(f"  0~5s   基准ACF RMS : "
      f"{np.std(sym_ref[lags_pos<=5]):.6f}")
print(f"\n  各天30~60s ACF RMS：")
print(f"  {'Day':<8} {'RMS':>10}  段落")
print("  " + "─"*30)
for i in range(N_DAYS):
    rms = np.std(daily_mat[(lags_pos>=30)&(lags_pos<=60), i])
    print(f"  Day{i+1:<4}  {rms:.6f}  {day_tag(i)}")
```

```python
# ══════════════════════════════════════════════════════════════════
# dv/v 计算（30天，全叠加参考）
# ══════════════════════════════════════════════════════════════════
from wxspectrum import wxspectrum, dvv_from_wxdt

N_DAYS     = 30
N_DAYS_REF = 10
N_DAYS_CUR = 20

dvv_true = np.zeros(N_DAYS)
dvv_true[10:15] = +TRUE_DVV
dvv_true[15:20] =  0.0
dvv_true[20:30] = -TRUE_DVV

seg_colors = (
    ['steelblue'] * 10 +
    ['tomato']    *  5 +
    ['seagreen']  *  5 +
    ['purple']    * 10
)

def day_tag(i):
    if i < 10: return '基准'
    if i < 15: return '增大'
    if i < 20: return '恢复'
    return '减小'

FILTER_MIN    = 1.0
FILTER_MAX    = 2.0
WX_WNAME      = 'amor'
WX_FREQ_LIMITS = (FILTER_MIN, FILTER_MAX)
WX_VOICES     = 10
WX_SMOOTHING  = False
WX_EXTEND     = True
DVV_FREQ_BAND = (FILTER_MIN, FILTER_MAX)
DVV_LAG_BAND  = (30, 60)
DVV_COI_MASK  = True

# 全叠加参考
x_ref_all = sym_all

dvv_syn    = np.full(N_DAYS, np.nan)
dvv_err_syn = np.full(N_DAYS, np.nan)

print(f"【全叠加参考】逐日dv/v计算...\n")
for i, corr_2d in enumerate(all_corr):
    fake_dict = {
        'corr'  : corr_2d,
        't'     : np.arange(corr_2d.shape[1], dtype=float),
        'lags'  : lags,
        'maxlag': float(np.max(np.abs(lags))),
    }
    s     = stack(fake_dict, method='linear', allstack=True)
    x_cur = get_pos(s['corr'][:, 0])
    try:
        wx  = wxspectrum(
            x_reference=x_ref_all, x_current=x_cur,
            fs=TARGET_SR, wname=WX_WNAME,
            freq_limits=WX_FREQ_LIMITS, smoothing=WX_SMOOTHING,
            voices_per_octave=WX_VOICES, extend_sig=WX_EXTEND,
            compute_coherence=False, time_axis=lags_pos,
        )
        res = dvv_from_wxdt(
            wxdt=wx['WXdt'], wxamp=wx['WXamp'],
            freqs=wx['freq'], time=wx['time'], coi=wx['coi'],
            freq_band=DVV_FREQ_BAND, lag_band=DVV_LAG_BAND,
            coi_mask=DVV_COI_MASK,
        )
        dvv_ts = res['dvv']
        wt_sum = res['weight'].sum(axis=0)
        valid  = np.isfinite(dvv_ts) & (wt_sum > 0)
        if valid.sum() > 0:
            wts          = wt_sum[valid]
            dv_mean      = np.average(dvv_ts[valid], weights=wts)
            wts_norm     = wts / wts.sum()
            dv_var       = np.sum(wts_norm * (dvv_ts[valid] - dv_mean)**2)
            n_eff        = wts.sum()**2 / (wts**2).sum()
            dv_std       = np.sqrt(dv_var * n_eff / max(n_eff-1, 1))
            dvv_syn[i]   = dv_mean
            dvv_err_syn[i] = dv_std
    except Exception as e:
        print(f"  Day{i+1} ❌ {e}")
    print(f"  Day{i+1:2d} ({day_tag(i)})  "
          f"dv/v={dvv_syn[i]:+.6f}%  err={dvv_err_syn[i]:.6f}%")


```

```python
# ══════════════════════════════════════════════════════════════════
# dv/v 计算（30天，纯基准段Day1~10参考）
# ══════════════════════════════════════════════════════════════════
def stack_days(corr_list, method='linear'):
    all_wins  = np.concatenate(corr_list, axis=1)
    fake_dict = {
        'corr'  : all_wins,
        't'     : np.arange(all_wins.shape[1], dtype=float),
        'lags'  : lags,
        'maxlag': float(np.max(np.abs(lags))),
    }
    result = stack(fake_dict, method=method, allstack=True)
    return result['corr'][:, 0]

stack_ref_baseline = stack_days(all_corr[:N_DAYS_REF])
x_ref_pure         = get_pos(stack_ref_baseline)

dvv_syn2     = np.full(N_DAYS, np.nan)
dvv_err_syn2 = np.full(N_DAYS, np.nan)

print(f"【纯基准段参考 Day1~10】逐日dv/v计算...\n")
for i, corr_2d in enumerate(all_corr):
    fake_dict = {
        'corr'  : corr_2d,
        't'     : np.arange(corr_2d.shape[1], dtype=float),
        'lags'  : lags,
        'maxlag': float(np.max(np.abs(lags))),
    }
    s     = stack(fake_dict, method='linear', allstack=True)
    x_cur = get_pos(s['corr'][:, 0])
    try:
        wx  = wxspectrum(
            x_reference=x_ref_pure, x_current=x_cur,
            fs=TARGET_SR, wname=WX_WNAME,
            freq_limits=WX_FREQ_LIMITS, smoothing=WX_SMOOTHING,
            voices_per_octave=WX_VOICES, extend_sig=WX_EXTEND,
            compute_coherence=False, time_axis=lags_pos,
        )
        res = dvv_from_wxdt(
            wxdt=wx['WXdt'], wxamp=wx['WXamp'],
            freqs=wx['freq'], time=wx['time'], coi=wx['coi'],
            freq_band=DVV_FREQ_BAND, lag_band=DVV_LAG_BAND,
            coi_mask=DVV_COI_MASK,
        )
        dvv_ts = res['dvv']
        wt_sum = res['weight'].sum(axis=0)
        valid  = np.isfinite(dvv_ts) & (wt_sum > 0)
        if valid.sum() > 0:
            wts           = wt_sum[valid]
            dv_mean       = np.average(dvv_ts[valid], weights=wts)
            wts_norm      = wts / wts.sum()
            dv_var        = np.sum(wts_norm * (dvv_ts[valid] - dv_mean)**2)
            n_eff         = wts.sum()**2 / (wts**2).sum()
            dv_std        = np.sqrt(dv_var * n_eff / max(n_eff-1, 1))
            dvv_syn2[i]   = dv_mean
            dvv_err_syn2[i] = dv_std
    except Exception as e:
        print(f"  Day{i+1} ❌ {e}")
    print(f"  Day{i+1:2d} ({day_tag(i)})  "
          f"dv/v={dvv_syn2[i]:+.6f}%  err={dvv_err_syn2[i]:.6f}%")

# ── 汇总统计 ──────────────────────────────────────────────────────
rmse1 = np.sqrt(np.nanmean((dvv_syn  - dvv_true*100)**2))
rmse2 = np.sqrt(np.nanmean((dvv_syn2 - dvv_true*100)**2))

print(f"\n{'='*60}")
print(f"真实dv/v：基准0% → 增大+{TRUE_DVV*100:.4f}% → 恢复0% → 减小-{TRUE_DVV*100:.4f}%")
print(f"\n{'段落':<20} {'期望':>8} {'全叠加参考':>12} {'纯基准参考':>12} {'恢复率(纯)':>10}")
print('─' * 66)
segs = [
    ('基准   Day1~10',  slice(0,10),    0.0),
    ('增大   Day11~15', slice(10,15), +TRUE_DVV*100),
    ('恢复   Day16~20', slice(15,20),   0.0),
    ('减小   Day21~30', slice(20,30), -TRUE_DVV*100),
]
for label, sl, expected in segs:
    m1 = np.nanmean(dvv_syn[sl])
    m2 = np.nanmean(dvv_syn2[sl])
    recovery = f'{abs(m2/expected)*100:.1f}%' if expected != 0 else 'N/A'
    print(f"  {label:<18} {expected:>+8.4f}% {m1:>+12.4f}% "
          f"{m2:>+12.4f}% {recovery:>10}")
print(f"\n  RMSE（全叠加参考）: {rmse1:.5f}%")
print(f"  RMSE（纯基准参考）: {rmse2:.5f}%")

# ── 可视化 ────────────────────────────────────────────────────────
days = np.arange(1, N_DAYS+1)

fig, axes = plt.subplots(2, 1, figsize=(16, 10), sharex=True)
fig.suptitle(
    f'合成数据dv/v验证（30天）  真实值：阶梯±{TRUE_DVV*100:.4f}%\n'
    f'lag={DVV_LAG_BAND}s  {FILTER_MIN}~{FILTER_MAX}Hz',
    fontweight='bold', fontsize=13
)

for ax, dvv_arr, dvv_err_arr, title in zip(
    axes,
    [dvv_syn,  dvv_syn2],
    [dvv_err_syn, dvv_err_syn2],
    [f'(a) 全叠加作参考  RMSE={rmse1:.5f}%',
     f'(b) 纯基准段Day1~10作参考  RMSE={rmse2:.5f}%'],
):
    # 段落背景
    for x0, x1, col_bg in [
        (1,  10, 'steelblue'),
        (11, 15, 'tomato'),
        (16, 20, 'seagreen'),
        (21, 30, 'purple'),
    ]:
        ax.axvspan(x0-0.5, x1+0.5, alpha=0.07, color=col_bg)

    # 真实值阶梯
    ax.step(days, dvv_true*100, where='mid',
            lw=2.5, color='k', label='真实dv/v', zorder=4)

    # 误差范围+计算值
    ax.fill_between(days,
                    dvv_arr - dvv_err_arr,
                    dvv_arr + dvv_err_arr,
                    alpha=0.2, color='gray')
    ax.plot(days, dvv_arr, 'o-', ms=5, lw=1.2, color='gray', zorder=2)
    for i in range(N_DAYS):
        ax.scatter(days[i], dvv_arr[i], color=seg_colors[i], s=60, zorder=3)

    ax.axhline( TRUE_DVV*100, color='tomato',  ls='--', lw=1.0,
                label=f'+{TRUE_DVV*100:.4f}%')
    ax.axhline( 0,            color='gray',    ls=':',  lw=0.8)
    ax.axhline(-TRUE_DVV*100, color='purple',  ls='--', lw=1.0,
                label=f'-{TRUE_DVV*100:.4f}%')
    ax.set_ylabel('dv/v (%)')
    ax.set_title(title)
    ax.legend(fontsize=8, loc='upper right', ncol=3)
    ax.grid(True, ls=':', alpha=0.4)

axes[-1].set_xlabel('Day')
axes[-1].set_xticks(days)
axes[-1].set_xticklabels([f'D{i}' for i in days], fontsize=8)

plt.tight_layout()
plt.savefig('synthetic_dvv_30days.png', dpi=130, bbox_inches='tight')
plt.show()
```

```python
from process_raw import remove_response
from obspy import read
from scipy.signal import butter, filtfilt
from scipy.fft import rfft, rfftfreq, next_fast_len
import numpy as np
import matplotlib.pyplot as plt
import glob

real_Z = sorted(glob.glob("../AL201/2019/*_Z_*.sac"))
print(f"Z分量文件数: {len(real_Z)}")
print(f"使用: {real_Z[0]}")

PAZ_QS10 = {
    'poles': [
        complex(-1.848e+02, 0.0),
        complex(-5.342e+00, 0.0),
        complex(-4.238e-01, 0.0),
        complex(-2.833e-01, 0.0),
    ],
    'zeros': [
        complex(-1.850e+02, 0.0),
        complex(-5.989e+00, 0.0),
        complex(0.0, 0.0),
        complex(0.0, 0.0),
    ],
    'gain'        : 7.778e+10,
    'sensitivity' : 6.271920e+08,
    'norm_freq'   : 0.2,          # ← 补上这个
}


tr_raw = read(real_Z[0])[0].copy()
fs     = tr_raw.stats.sampling_rate
x_raw  = tr_raw.data.astype(np.float64)
print(f"\n台站   : {tr_raw.id}")
print(f"采样率 : {fs} Hz")
print(f"原始max: {np.max(np.abs(x_raw)):.4e} counts")
print(f"原始std: {np.std(x_raw):.4e} counts")

# 方法1：你的remove_response
x_yours = remove_response(x_raw.copy(), fs, PAZ_QS10,
                           PRE_FILT, water_level_db=60.0)
print(f"\n你的方法  max={np.max(np.abs(x_yours)):.4e} m/s  "
      f"std={np.std(x_yours):.4e} m/s")

# 方法2：obspy simulate
tr_obspy = tr_raw.copy()
tr_obspy.simulate(
    paz_remove   = PAZ_QS10,
    paz_simulate = None,
    pre_filt     = PRE_FILT,
    water_level  = 60.0,
)
x_obspy = tr_obspy.data
print(f"obspy方法 max={np.max(np.abs(x_obspy)):.4e} m/s  "
      f"std={np.std(x_obspy):.4e} m/s")

N     = min(len(x_yours), len(x_obspy), int(3600*fs))
ratio = np.std(x_obspy[:N]) / (np.std(x_yours[:N]) + 1e-30)
print(f"\n比值(obspy/yours) = {ratio:.6f}  "
      f"{'✅' if 0.9 < ratio < 1.1 else '⚠️ 差异过大'}")

b01, a01 = butter(4, [0.1, 10.0], btype='band', fs=fs)
y1 = filtfilt(b01, a01, x_yours[:N])
y2 = filtfilt(b01, a01, x_obspy[:N])
cc = np.corrcoef(y1, y2)[0, 1]
print(f"相关系数(0.1~10Hz) = {cc:.8f}  "
      f"{'✅' if cc > 0.999 else '⚠️ 不一致'}")

b12, a12 = butter(4, [1.0, 2.0], btype='band', fs=fs)
z1 = filtfilt(b12, a12, x_yours[:N])
z2 = filtfilt(b12, a12, x_obspy[:N])
cc12 = np.corrcoef(z1, z2)[0, 1]
print(f"相关系数(1~2Hz)    = {cc12:.8f}  "
      f"{'✅' if cc12 > 0.999 else '⚠️ 不一致'}")

rms_yours = np.std(x_yours[:N])
rms_obspy = np.std(x_obspy[:N])
print(f"\n量级检验：")
print(f"  你的方法  RMS={rms_yours:.4e} m/s  "
      f"{'✅' if 1e-9 < rms_yours < 1e-4 else '⚠️'}")
print(f"  obspy方法 RMS={rms_obspy:.4e} m/s  "
      f"{'✅' if 1e-9 < rms_obspy < 1e-4 else '⚠️'}")

n_show = min(int(300*fs), N)
t_show = np.arange(n_show) / fs

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle(f'去仪器响应方法对比  {tr_raw.id}  CC={cc:.6f}',
             fontweight='bold')

ax = axes[0, 0]
ax.plot(t_show, y1[:n_show]/(np.std(y1)+1e-30) + 3,
        lw=0.5, color='steelblue', label='你的方法（+3）')
ax.plot(t_show, y2[:n_show]/(np.std(y2)+1e-30),
        lw=0.5, color='tomato',    label='obspy方法', alpha=0.8)
ax.set_xlabel('Time (s)')
ax.set_title(f'(a) 波形对比（0.1~10Hz，前300s）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.4)

ax   = axes[0, 1]
diff = y1[:n_show]/(np.std(y1)+1e-30) - y2[:n_show]/(np.std(y2)+1e-30)
ax.plot(t_show, diff, lw=0.5, color='purple')
ax.axhline(0, color='k', ls='--', lw=0.8)
ax.set_xlabel('Time (s)')
ax.set_title(f'(b) 差值  max={np.max(np.abs(diff)):.4e}')
ax.grid(True, ls=':', alpha=0.4)

Nfft = next_fast_len(N)
fq   = rfftfreq(Nfft, d=1.0/fs)
sp1  = np.abs(rfft(x_yours[:N], Nfft))
sp2  = np.abs(rfft(x_obspy[:N], Nfft))

ax = axes[1, 0]
ax.loglog(fq[1:], sp1[1:]/(np.max(sp1[1:])+1e-30),
          lw=0.8, color='steelblue', label='你的方法', alpha=0.8)
ax.loglog(fq[1:], sp2[1:]/(np.max(sp2[1:])+1e-30),
          lw=0.8, color='tomato',    label='obspy方法', alpha=0.8)
ax.axvspan(1.0, 2.0, alpha=0.1, color='green', label='目标1~2Hz')
ax.set_xlim(0.01, 25)
ax.set_xlabel('Frequency (Hz)')
ax.set_title('(c) 频谱对比（归一化）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.3)

ax = axes[1, 1]
ax.plot(t_show, z1[:n_show]/(np.std(z1)+1e-30) + 3,
        lw=0.6, color='steelblue',
        label=f'你的方法（+3）  CC={cc12:.6f}')
ax.plot(t_show, z2[:n_show]/(np.std(z2)+1e-30),
        lw=0.6, color='tomato', label='obspy方法', alpha=0.8)
ax.set_xlabel('Time (s)')
ax.set_title('(d) 1~2Hz波形对比（前300s）')
ax.legend(fontsize=8)
ax.grid(True, ls=':', alpha=0.4)

plt.tight_layout()
plt.savefig('response_compare.png', dpi=130, bbox_inches='tight')
plt.show()

print(f"\n{'='*50}")
print(f"汇总：")
print(f"  振幅比值(obspy/yours) = {ratio:.6f}  "
      f"{'✅' if 0.9<ratio<1.1 else '⚠️'}")
print(f"  相关系数(0.1~10Hz)   = {cc:.8f}  "
      f"{'✅' if cc>0.999 else '⚠️'}")
print(f"  相关系数(1~2Hz)      = {cc12:.8f}  "
      f"{'✅' if cc12>0.999 else '⚠️'}")
print(f"  量级(yours)          = {rms_yours:.4e} m/s  "
      f"{'✅' if 1e-9<rms_yours<1e-4 else '⚠️'}")
if cc > 0.999 and 0.9 < ratio < 1.1 and 1e-9 < rms_yours < 1e-4:
    print(f"\n  ✅ 你的remove_response实现正确")
else:
    print(f"\n  ⚠️ 存在问题，建议改用obspy simulate")
```
